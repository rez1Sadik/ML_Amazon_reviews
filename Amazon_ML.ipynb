# **Sentiment Analysis using Machine Learning for Amazon Product Reviews**

## **Data Preprocessing**

**Objective:**
<br>The objective of this data preprocessing and sentiment analysis script is to prepare a dataset of product reviews for sentiment analysis, particularly focusing on English-language reviews. The sentiment analysis aims to classify reviews into 'positive' or 'negative' sentiments based on star ratings.



### **Stage-1:**

1. **Loading Data:**
   - Read the CSV file ('AllProductReviews.csv') into a pandas DataFrame named 'data_amazon.'
   - Drop rows with NaN values in the 'ReviewStar,' 'ReviewBody,' and 'Product' columns.

2. **Filtering Reviews:**
   - Remove rows with a star value of 3, as neutral reviews might not contribute to sentiment analysis effectively.

3. **Text Cleaning:**
   - Define a function 'clean_text' to perform the following operations on each review:
     - Expand contractions using the 'contractions' library.
     - Remove HTML tags and URLs.
     - Remove non-alphanumeric characters.
     - Remove emoticons.
     - Remove extra white spaces and leading/trailing spaces.
     - Convert text to lowercase.
     - Add space after numbers followed by letters.
     - Replace underscores with spaces.
   - Apply 'clean_text' function to the 'ReviewBody' column with a progress bar.

4. **Language Filtering:**
   - Define a function 'is_english' to check if the language of a review is English, using the 'langdetect' library.
   - Drop rows where the language is not English, applying the 'is_english' function with a progress bar.

5. **Additional Data Cleaning:**
   - Drop rows with empty 'Review_text.'
   - Drop rows with NaN values in the 'ReviewStar' and 'Review_text' columns.

6. **Sentiment Labeling:**
   - Create a new column 'sentiment' based on star values. Assign 'positive' to reviews with star ratings 4 and 5, and 'negative' to reviews with other ratings.

7. **Column Selection:**
   - Keep only the relevant columns ('Review_text', 'sentiment', 'Product', 'ReviewBody') in the DataFrame.

8. **Result Display:**
   - Display the modified DataFrame with the selected columns.

### **Stage-2:**

1. **Brand Name Removal:**
   - Define a list named 'words_to_remove' containing brand names and model numbers to be removed from the reviews.

2. **Review Text Modification:**
   - Define a function named 'remove_brand' to remove specified brand names and model numbers from a given review.
     - Use regular expressions for case-insensitive partial matching to avoid removing words unintentionally.
     - Apply the 'remove_brand' function to the 'Review_text' column in the 'data_amazon' DataFrame.



### **Stage-3:**

1. **Word Replacement Function:** (the list is collected by using misspelled word frequency[count per review])
   - Define a function named 'replace_awe_words' to replace words starting with 'awe' with 'awesome' in a given text.
     - Use the 're' (regular expression) module to perform the replacements.
     - Substitutes words starting with 'fan' with 'fantastic.'
     - Substitutes words starting with 'awe' with 'awesome.'
     - Substitutes words starting with 'excel' or 'xcell' with 'excellent.'
     - Substitutes words starting with 'perfe' with 'perfect.'
     - Substitutes words starting with 'wonde' with 'wonderful.'
     - Returns the modified text.

2. **Application of Replacement Function:**
    - Apply the 'replace_awe_words' function to the 'Review_text' column of the 'data_amazon' DataFrame.
    - Update the 'Review_text' column with the modified text.

### **Stage-4:**

1. **Importing Libraries:**
   - Import necessary libraries for text preprocessing, sentiment analysis, and data manipulation, including `re`, `nltk`, `pandas`, `stopwords`, `WordNetLemmatizer`, `SentimentIntensityAnalyzer` from `nltk.sentiment`, and `tqdm`.

2. **Downloading NLTK Resources:**
   - Download required NLTK resources such as stopwords, punkt tokenizer, and VADER lexicon for sentiment analysis.

3. **Stopwords and Lemmatization Initialization:**
   - Load English stopwords and initialize a lemmatizer from the WordNetLemmatizer class.
   - Remove 'not' and 'ok' from the stopwords set.

4. **Sentiment Intensity Analyzer Initialization:**
   - Initialize a SentimentIntensityAnalyzer (SIA) for sentiment analysis using VADER.

5. **Text Preprocessing:**
   - Remove stopwords from the 'Review_text' column of the 'data_amazon' DataFrame using tqdm for progress tracking.
   - Perform lemmatization on the processed text.

6. **Drop Rows with NaN Values:**
   - Drop rows with NaN values in the 'sentiment' and 'Review_text_n' columns, creating a new DataFrame 'data_amazon_1.'

7. **Function to Check Positive or Negative Words:**
   - Define a function named 'contains_positive_or_negative' that takes a review as input and checks if it contains at least one positive or negative word using VADER.

8. **Filtering Based on Conditions:**
   - Apply the 'contains_positive_or_negative' function to filter reviews. Keep reviews containing positive or negative words.

9. **Save Dropped Reviews:**
   - Save dropped reviews (not meeting conditions) in a separate DataFrame named 'dropped_reviews_amazon.'

10. **Result Display:**
    - Display the preprocessed DataFrame 'filtered_data_amazon' containing reviews that meet the specified conditions.

**Conclusion:**
<br>The resulting 'data_amazon' DataFrame is now ready for sentiment analysis tasks, with cleaned and preprocessed text data and sentiment labels assigned based on star ratings. The script ensures that the dataset is free from unnecessary or non-English reviews, providing a clean and focused dataset for subsequent analysis.
